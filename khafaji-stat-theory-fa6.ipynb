```python
grades = ["90-100", "80-89","70-79", "60-69","50-59", "40-49", "30-39"]
freq = [9, 32, 43, 21, 11, 3, 1]
freq.reverse()

class_lowerbound = [90, 80, 70, 60, 50, 40, 30]
class_lowerbound.reverse()

class_size = [100-90, 89-80, 79-70, 69-60, 59-50, 49-40, 39-30]
class_size.reverse()

freqnp = np.array(freq)
cumfreq = np.cumsum(freqnp)
```

```python
def quartiles(quartile, freq, cumfreq, lowerbounds, class_size):
    
    in_dividedby_4 = (quartile*sum(freq))/4
    
    least_interpol = [abs(x-in_dividedby_4) for x in cumfreq]
    
    least_interpol_min = min(least_interpol)
    
    index_used = least_interpol.index(least_interpol_min)
    
    quartile_res = lowerbounds[index_used] + (class_size[index_used]/freq[index_used]) * (in_dividedby_4 - cumfreq[index_used])
    
    print("The ", quartile,"th quartile is: ", quartile_res)
```

```python
quartiles(1, freq, cumfreq, class_lowerbound, class_size)
quartiles(2, freq, cumfreq, class_lowerbound, class_size)
quartiles(3, freq, cumfreq, class_lowerbound, class_size)
```

```python
sett = [6,2,8,7,5]
mean = sum(sett)/len(sett)
stdev = statistics.stdev(sett)

standardized = [(x-mean)/stdev for x in sett]

mean_standard = sum(standardized)/len(standardized)

stdev_standard = statistics.stdev(standardized)

print("The mean of a standardized data set is: ", mean_standard," (infinitesmally small that it is practically 0) and its standard deviation is: ", stdev_standard )

```

```python
x_sett_5 = [6, 9, 12, 15, 18]
px_sett_5 = [0.1, 0.2, 0.4, 0.2, 0.1]

summ = 0
for x, px in zip(x_sett_5, px_sett_5):
    summ += x*px

mean_5 = summ/len(x_sett_5)

summ_x_mean_squared = 0 

for x, px in zip(x_sett_5, px_sett_5):
    summ_x_mean_squared += (x*px - mean_5)**2
    
variance_5 = summ_x_mean_squared/len(x_sett_5)

samples_size2_5 = list(itertools.product(x_sett_5, x_sett_5)) 

mean_products = [ sum(x)/2 for x in samples_size2_5]

probab_products = []

for tup in samples_size2_5:
    probabbi = 1
    for num in tup:
        probabbi = probabbi * px_sett_5[x_sett_5.index(num)]
    probab_products.append(probabbi)
    
print(probab_products)
```

```python
table_data = {
    "Product" : samples_size2_5,
    "mean" : mean_products,
    "probability" : probab_products
}

df = pl.DataFrame(table_data)

with pl.Config(tbl_rows=25):

    print(df)
```
