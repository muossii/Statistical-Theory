```python
responses = ["Marcos", "Marcos", "Robredo", "Robredo", "Marcos", "Marcos"]

r = 4

import itertools

samples = list(itertools.combinations(responses, r))

```

```python
samples
```

```python
dict_key = "Sample"

cols = dict()
for count, sample in enumerate(samples):
    cols[dict_key + str(count+1)] = sample
    
cols
```

```python
proportion_robredo_samples = [ sample.count("Robredo")/len(sample) for sample in samples]

print("Robredo samples proportions: ",proportion_robredo_samples)

proportion_marcos_samples = [ sample.count("Marcos")/len(sample) for sample in samples]

print("Marcos samples proportions: ",proportion_marcos_samples)
```

```python
import polars as pl

table = pl.LazyFrame(cols)

table.collect()
```

```python
column_names = table.collect_schema().names()

column_names
```

```python
prop = {col : [str(prop_mar), str(prop_rob)] for col, prop_mar, prop_rob in zip(column_names,proportion_marcos_samples, proportion_robredo_samples)
    
}

prop
    

```

```python
prop_lf = pl.LazyFrame(prop)

prop_lf.collect()
```

```python
full_lf = pl.concat([table, prop_lf])

full_lf = full_lf.with_columns(pl.Series("Elections", ["Voter 1", "Voter 2", "Voter 3", "Voter 4", "Marcos Sample Proportion", "Robredo Sample Proportion"])
)

full_lf.collect()
```

```python
col_tochange = full_lf.collect_schema().names()

col_tochange[0], col_tochange[15] = col_tochange[15], col_tochange[0]

full_lf.select(col_tochange).collect()
```

```python
import statistics

proportion_robredo_samples_ave = statistics.mean(proportion_robredo_samples)

print("mean of samples proportions of Marcos is: ", proportion_robredo_samples_ave)

print("\n")

proportion_marcos_samples_ave = statistics.mean(proportion_marcos_samples)

print("mean of samples proportions of Marcos is: ", proportion_marcos_samples_ave)
```
